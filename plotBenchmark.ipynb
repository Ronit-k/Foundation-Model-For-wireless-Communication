```python
import matplotlib.pyplot as plt

split_ratio = [0.2, 0.4, 0.6, 0.8]

# Data
base_cnn_cls = [0.3885, 0.4666, 0.5182, 0.5326]
base_cnn_channel = [0.8603, 0.8931, 0.9094, 0.9171]

secnn_base_cls = [0.4623, 0.5409, 0.5773, 0.6030]
secnn_base_channel = [0.8909, 0.9104, 0.9228, 0.9290]

secnn_ca_cls = [0.6653, 0.7328, 0.7573, 0.7837]
secnn_ca_channel = [0.9053, 0.9195, 0.9268, 0.9354]

plt.figure(figsize=(10, 6))

# ----- CA-LWM + SE-CNN (Dark colors) -----
line_dark_green, = plt.plot(
    split_ratio, secnn_ca_channel,
    marker='o', color='#006400',  # dark green
    label='CA-LWM + SE-CNN (channel_emb)'
)

line_dark_blue, = plt.plot(
    split_ratio, secnn_ca_cls,
    marker='o', color='#00008B',  # dark blue
    label='CA-LWM + SE-CNN (cls_emb)'
)

# ----- Base LWM + SE-CNN (Medium colors — lighter than dark, darker than light) -----
line_green, = plt.plot(
    split_ratio, secnn_base_channel,
    marker='o', color='#2ca02c',  # medium green
    label='Base LWM + SE-CNN (channel_emb)'
)

line_blue, = plt.plot(
    split_ratio, secnn_base_cls,
    marker='o', color='#1f77b4',  # medium blue
    label='Base LWM + SE-CNN (cls_emb)'
)

# ----- Base LWM + Base CNN (Light colors) -----
line_light_green, = plt.plot(
    split_ratio, base_cnn_channel,
    marker='o', color='#98df8a',  # light green
    label='Base LWM + Base CNN (channel_emb)'
)

line_light_blue, = plt.plot(
    split_ratio, base_cnn_cls,
    marker='o', color='#9ecae1',  # light blue
    label='Base LWM + Base CNN (cls_emb)'
)

plt.xlabel("Training Split Ratio")
plt.ylabel("Weighted F1 Score")

plt.title(
    "Benchmark for Beam Prediction Task:\n"
    "CA-LWM with Improved SE-CNN vs Base LWM Configurations"
)

plt.grid(True, linestyle='--', alpha=0.6)

# ===== Legend order you requested =====
plt.legend(handles=[
    line_dark_green,
    line_green,
    line_light_green,
    line_dark_blue,
    line_blue,
    line_light_blue
], loc='lower right')

plt.tight_layout()
plt.show()
```
